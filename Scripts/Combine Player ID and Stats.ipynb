```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
stats = pd.read_csv("NFL_2010-2020_Data.csv")
info = pd.read_csv("Player_Info - Player_Info.csv")
```

```python
df = stats
df2 = info
```

```python
df['name'] = df.player_id.map(df2.set_index('player_id')['name'].to_dict())
df['position'] = df.player_id.map(df2.set_index('player_id')['position'].to_dict())
```

```python
df3 = df[df['name'].notna()]
```

```python
df3
```

```python
first_column = df3.pop('name')
df3.insert(0, 'Name', first_column)
Second_column = df3.pop('position')
df3.insert(1, 'Position', Second_column)
```

```python
df3 = df3.drop(['fantasy_points', 'fantasy_points_ppr','ppr_sh','dakota'], axis=1)
```

```python
Stats_Names = df3
```

```python
cnx = sql.connect("Stats_Names.db") #input Database Name
```

```python
Stats_Names.to_sql('Stats_with_names-2', cnx, if_exists='replace', index = False) # Exports data to an SQL Lite Database
```

```python
cnx.close()
```
