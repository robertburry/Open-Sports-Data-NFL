# Dependencies

```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

# Open SQL Connection

```python
db_stats_cnx = sql.connect('../Data/Yearly Data as Database/NFL_2021_Data.db')
db_player_cnx = sql.connect('../Data/Yearly Data as Database/Player_Info.db')
cnx = sql.connect("Stats_with_names_2021.db") 
```

# Combining Databases

```python
#stats = pd.read_csv("NFL_2010-2020_Data.csv")
#info = pd.read_csv("Player_Info - Player_Info.csv")

# Reading data in from the databases as opposed to the csv files
stats = pd.read_sql_query('SELECT * FROM NFL_2021_Data', db_stats_cnx)
info = pd.read_sql_query('SELECT * FROM Player_Info', db_player_cnx)
```

```python
stats['player_id'] = stats['player_id'].str.replace('-','')
```

```python
stats['player_id'] = pd.to_numeric(stats['player_id'])
```

```python
stats['name'] = stats.player_id.map(info.set_index('player_id')['name'].to_dict())
stats['position'] = stats.player_id.map(info.set_index('player_id')['position'].to_dict())
```

```python
stats
```

```python
combined = stats[stats['name'].notna()]
```

```python
combined
```

```python
first_column = combined.pop('name')
combined.insert(0, 'Name', first_column)
Second_column = combined.pop('position')
combined.insert(1, 'Position', Second_column)
```

```python
combined = combined.drop(['fantasy_points', 'fantasy_points_ppr','ppr_sh','dakota'], axis=1)
```

```python
Stats_Names = combined
```

```python
Stats_Names
```

# Writing Dataframe to SQL Database

```python
Stats_Names.to_sql('Stats_with_names_2021', cnx, if_exists='replace', index = False) # Exports data to an SQL Lite Database
```
