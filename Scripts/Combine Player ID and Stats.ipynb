```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
db_stats = sql.connect('../Data/Yearly Data as Database/NFL_2010-2020_Data.db') #input Database Name
db_player = sql.connect('../Data/Yearly Data as Database/Player_Info.db')
```

```python
#stats = pd.read_csv("NFL_2010-2020_Data.csv")
#info = pd.read_csv("Player_Info - Player_Info.csv")

# Reading data in from the databases as opposed to the csv files
stats = pd.read_sql_query('SELECT * FROM NFL_2010_2020_Data', db_stats)
info = pd.read_sql_query('SELECT * FROM Player_Info', db_player)
```

```python
stats['name'] = stats.player_id.map(info.set_index('player_id')['name'].to_dict())
stats['position'] = stats.player_id.map(info.set_index('player_id')['position'].to_dict())
```

```python
combined = stats[stats['name'].notna()]
```

```python
combined
```

```python
first_column = combined.pop('name')
combined.insert(0, 'Name', first_column)
Second_column = combined.pop('position')
combined.insert(1, 'Position', Second_column)
```

```python
combined = combined.drop(['fantasy_points', 'fantasy_points_ppr','ppr_sh','dakota'], axis=1)
```

```python
Stats_Names = combined
```

```python
Stats_Names
```

```python
cnx = sql.connect("Stats_Names.db") #input Database Name
```

```python
Stats_Names.to_sql('Stats_with_names', cnx, if_exists='replace', index = False) # Exports data to an SQL Lite Database
```
